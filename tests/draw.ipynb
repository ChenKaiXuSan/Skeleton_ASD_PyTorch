```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 
plt.rcParams.update({'font.size': 30, 'font.family': 'sans-serif'})

# 
confusion_matrix_data = np.array([[0.4074, 0.5926], [0.3704, 0.6296]])

axis_labels = ['ASD', 'non_ASD']

# matplotlibseaborn
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix_data, annot=True, fmt='.4f', cmap='Reds', xticklabels=axis_labels, yticklabels=axis_labels, vmin=0, vmax=1, annot_kws={'color': 'black'})
plt.title('Doctor 1 result', fontsize=30)
plt.ylabel('Actual Label', fontsize=30)
plt.xlabel('Predicted Label', fontsize=30)
plt.show()

```
